### Parking
From Smart Flanders, we were able to get historical parking data of about 1 year of Leuven. It is one big file which contains, in csv format, the occupation level of each underground parking garage, measured about every 30 seconds. 
Source: https://datapiloten.be/parking/ . Fun fact, if you replace `leuven.csv` in the download cell (all the way on top) with `gent.csv`, `kortrijk.csv` or `sint-niklaas.csv` you get the data of those cities. 

```python
import pandas as pd
df = pd.read_csv('data/parking/leuven.csv', sep=',')
print(df.columns)
#df
```

```python
df['city'] = 'Leuven'
df['timestamp'] = pd.to_datetime(df['timestamp'], format="%Y/%m/%d %H:%M:%S")
last_month = df[(df['timestamp'] >= '2018-05-01 00:00:00') & (df['timestamp'] < '2018-06-01 00:00:00')]
last_month['dow']=last_month['timestamp'].dt.dayofweek
last_month['date']=last_month['timestamp'].dt.date
```

```python
aggregated = (last_month
        .groupby([last_month['timestamp'].dt.hour, last_month['parking'], last_month['date']]) #df['parking'], d
        .aggregate({'spaces': 'mean'}).reset_index())
aggregated['dow']=df['timestamp'].dt.dayofweek
aggregated
```

```python
%matplotlib inline
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.tsplot(data=aggregated, time='timestamp', condition='parking', unit='date', value='spaces')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ladeuze = (df[(df['timestamp'] >= '2017-01-01 00:00:00') &
            (df['timestamp'] < '2018-06-01 00:00:00') &
             (df['parking'] == 'Ladeuze')])
ladeuze['date']=ladeuze['timestamp'].dt.date
ladeuze = (ladeuze.groupby([ladeuze['timestamp'].dt.hour, ladeuze['parking'], ladeuze['date']]) #df['parking'], d
    .aggregate({'spaces': 'mean'}).reset_index())
ladeuze
ax = sns.tsplot(data=ladeuze, time='timestamp', unit='date', value='spaces', err_style="unit_traces")
```

```python
import numpy as np
import matplotlib.pyplot as plt
ladeuze['dow'] = pd.to_datetime(ladeuze['date'], format="%Y-%m-%d").dt.dayofweek
grouped = ladeuze.groupby('dow')
ncols=2
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20,15), sharey=True)

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
    sns.tsplot(data=grouped.get_group(key),  ax=ax, condition='dow', time='timestamp', unit='date', value='spaces', err_style="unit_traces")

ax.legend()
plt.show()
```

### Airquality
Downloaded from http://archive.luftdaten.info/, this is an international repository of air quality measurements. For the hackathon, we've only kept the measurements within a radius of 5km of Leuven. If you want to work on all the sensors, see the Downloads notebook. Warning: filesizes are big. 

```python
import pandas as pd
df = pd.read_csv('data/airquality/2018-01-30.csv', sep=',')
print(df.columns)
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], format="%Y-%m-%dT%H:%M:%S")
df['dow']=df['timestamp'].dt.dayofweek
df['time']=df['timestamp'].dt.time
df['date'] = df['timestamp'].dt.date
df_p1 = df[df['P1'] > 0]
print(df_p1.describe())
df_p1_aggregated = (df_p1
                    .groupby([df['timestamp'].dt.hour, df['sensor_id'], df['date']]) #df['parking'], d
                    .aggregate({'P1': 'mean', 'P2': 'mean'}).reset_index())
df_p1_aggregated = df_p1_aggregated[df_p1_aggregated["P1"] <= 50]
df_p1_aggregated
```

```python
%matplotlib inline
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.tsplot(data=df_p1_aggregated, time='timestamp', unit='sensor_id', value='P1', err_style="unit_traces")
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
